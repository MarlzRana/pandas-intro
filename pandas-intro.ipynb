A program that print the first 10 numbers of the Fibonacci sequence

```python
def fib_number_for_a_given_index(n):
    if n <= 1:
        return n
    return fib_number_for_a_given_index(n-1) + fib_number_for_a_given_index(n-2)

for i in range(1,11):
    print(fib_number_for_a_given_index(i))
```

Use the skiprows parameter when reading the csv to get rid of the first row and push it up as the column, so that column has now labels and not meaningless column index's

```python
olympicsdf = pd.read_csv("olympics.csv", skiprows=[0])
print(olympicsdf.columns)
olympicsdf.head()
```

The way medals are labelled within the columns are not intuitive. 

Let's label them in a better way using either:
- df.rename(columns={}) where we pass in a dictionary like:
  - {"previousColumnName": "newColumnName"}
  - This returns a new dataframe, the original dataframe is unmodified
- df.column_list = new_column_list where we must ensure that len(column_list) = len(new_column_list)

I am going to use .rename as we are not replacing the entire column_set

```python
olympicsdf = olympicsdf.rename(columns={
    "01 !" : "GS",
    "02 !": "SS",
    "03 !": "BS",
    "Total": "TS",
    "01 !.1": "GW",
    "02 !.1": "SW",
    "03 !.1": "BW",
    "Total.1": "TW",
    "01 !.2": "GT",
    "02 !.2": "ST",
    "03 !.2": "BT"
})
olympicsdf.head()
```

Find where the current Olympics host (China) ranks in the leaderboard of countries with the most gold medals in the winter Olympics

```python
df = olympicsdf.copy()
df["GWRank"] = df["GW"].rank(ascending=False, method="dense")
print("Rank : " + str(int(df.loc[df["Unnamed: 0"] == "China (CHN) [CHN]"]["GWRank"])))
```

Creating a dataframe from a dictionary

Creating a dataframe using the given table in the lab manual

```python
vinyl_dict = {
    "Album Name": ["The Stone Roses", "Definitely Maybe", "The Queen Is Dead"],
    "Band Name": ["The Stone Roses", "Oasis", "The Smiths"], 
    "Year": [1989, 1994, 1986],
}

vinyl_df = pd.DataFrame.from_dict(vinyl_dict)
vinyl_df
```
